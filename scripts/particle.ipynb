```python
import pandas as pd
from ase import Atoms
from ase.visualize import view
from ase.visualize.plot import plot_atoms
from autoadsorbate import Surface
from autoadsorbate.Particle import random_points_in_sphere
```

```python
particle = random_points_in_sphere(5, 40, center=(0, 0, 0))
particle_atoms = Atoms(['Pd']*len(particle), particle)

s = Surface(
    particle_atoms,
    mode='particle',
    precision=1.
    )
```

```python
view_atoms = s.atoms.copy()

for i in s.site_df.index.values:
    site = s.view_site(i, return_atoms=True)
    view_atoms += site[[atom.index for atom in site if atom.symbol == 'X']]
# view(view_atoms)

plot_atoms(view_atoms)
```
